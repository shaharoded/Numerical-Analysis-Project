```python
def bezier_coefficient_matrix(n):
  def ncr(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer // denom
  M = np.zeros((n+1,n+1), dtype=np.float64)
  for i in range(n + 1):
    for j in range(n + 1):
      if i > j:
        M[i][n-j] = 0
      else:
        M[i][n-j] = ((-1)**(j-i))*(ncr(n, i))*(ncr(n-i, j-i))
  return M
```

```python
def forward_sub(lowwer, b):
    n = lowwer.shape[0]
    y = np.zeros_like(b, dtype=np.double)
    y[0] = b[0] / lowwer[0, 0]
    for i in range(1, n):
        y[i] = (b[i] - np.dot(lowwer[i, :i], y[:i])) / lowwer[i, i]
    return y

def back_sub(upper, y):
    n = upper.shape[0]
    x = np.zeros_like(y, dtype=np.double)
    x[-1] = y[-1] / upper[-1, -1]
    for i in range(n - 2, -1, -1):
        x[i] = (y[i] - np.dot(upper[i, i:], x[i:])) / upper[i, i]
    return x

def lowwer_upper(matrix):
    n = matrix.shape[0]
    U = matrix.copy()
    L = np.eye(n, dtype=np.double)
    P = np.eye(n, dtype=np.double)
    for i in range(n):
        for k in range(i, n):
            if ~np.isclose(U[i, i], 0.0):
                break
            U[[k, k + 1]] = U[[k + 1, k]]
            P[[k, k + 1]] = P[[k + 1, k]]
        factor = U[i + 1:, i] / U[i, i]
        L[i + 1:, i] = factor
        U[i + 1:] -= factor[:, np.newaxis] * U[i]
    return P, L, U

def inverse(A):
    n = A.shape[0]
    b = np.eye(n)
    invers_matrix = np.zeros((n, n))
    P, L, U = lowwer_upper(A)
    for i in range(n):
        y = forward_sub(L, np.dot(P, b[i, :]))
        invers_matrix[i, :] = back_sub(U, y)
    return invers_matrix
```

```python
def bezier_curve_iterative(points, degree):
    """
    points : array of shape (n,2) containing n control points of bezier curve.
    degree : degree of the bezier curve 
    """
    def f(t):
        current_points = points
        for _ in range(degree):
            next_points = np.array([(1-t)*current_points[i] + t*current_points[i+1] for i in range(current_points.shape[0]-1)])
            current_points = next_points
        return current_points[0]
    return f
```

```python
def NOISY(noise):
    def decorate(function):
        def wrapper(*args, **kwargs):
            result = function(*args, **kwargs)
            result += np.random.randn() * noise
            return result

        return wrapper

    return decorate
```

```python
def inverse_1(A):
    """
    Inverses a given matrix A
    :param A: An invertible Matrix
    :return: The inversed matrix
    """
    # Check that the matrix is square
    if A.shape[0] != A.shape[1]:
      raise ValueError("Matrix is not square")
    n = len(A)
    AM = np.matrix.copy(A)
    I = np.identity(n)
    IM = np.matrix.copy(I)
    indices = list(range(n))
    for fd in range(n):
        fdScaler = 1.0 / AM[fd][fd]
        for j in range(n):
            AM[fd][j] *= fdScaler
            IM[fd][j] *= fdScaler
        for i in indices[0:fd] + indices[fd+1:]:
            crScaler = AM[i][fd]
            for j in range(n):
                AM[i][j] = AM[i][j] - crScaler * AM[fd][j]
                IM[i][j] = IM[i][j] - crScaler * IM[fd][j]
    return IM
```

Using a recursive algoritm to get the value of the function for t

```python
def bezier_curve_rec(points, degree, t):  # Using de Casteljau's algorithm
    """
    points : array of shape (n,2) containing n control points of bezier curve.
    degree : degree of the bezier curve 
    t      : parameter at which the point of bezier curve to be evaluated
    """
    if degree == 0:
        return points[0]
    else:
        next_points = np.array([(1-t)*points[i] + t*points[i+1] for i in range(degree)])
        return bezier_curve_rec(next_points, degree-1, t)
```

```python
def get_control_points(points, d):
    n = len(points)
    A = np.zeros((n, d + 1))
    for i, (x, y) in enumerate(points):
        for j in range(d + 1):
            A[i][j] = x ** j
    Y = np.array([y for x, y in points])
    # Solve the linear system A*P = Y using Gaussian elimination
    n = d + 1
    for k in range(n):
        for i in range(k + 1, n):
            if A[i, k] != 0:
                lambda_ = A[i, k] / A[k, k]
                A[i, k + 1 : n] = A[i, k + 1 : n] - lambda_ * A[k, k + 1 : n]
                Y[i] = Y[i] - lambda_ * Y[k]
    # Back substitution
    P = np.zeros(n)
    for i in range(n - 1, -1, -1):
        P[i] = (Y[i] - np.dot(A[i, i + 1 : n], P[i + 1 : n])) / A[i, i]
    # reshape the output
    control_points = []
    for i in range(d + 1):
        control_points.append([i, P[i]])
    return control_points
```

```python
def least_squares_approximation(d, points):
    n = len(points)
    # Create the design matrix A
    A = np.zeros((n, d+1))
    for i in range(n):
        for j in range(d+1):
            A[i][j] = points[i][0]**j
    # Create the data vector b
    b = np.array([points[i][1] for i in range(n)])
    # Solve the linear system A*x = b using custom inverse algorithm
    x = np.dot(np.dot(inverse(np.dot(A.T, A)), A.T), b)
    return x
```

```python
def bezier_coefficient_matrix(n):
    M = np.zeros((n+1,n+1), dtype=np.float64)
    for i in range(n + 1):
      for j in range(n + 1):
        if i > j:
          M[i][n-j] = 0
        else:
          M[i][n-j] = ((-1)**(j-i))*(ncr(n, i))*(ncr(n-i, j-i))
    return M


import operator as op
from functools import reduce

def ncr(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer // denom
```

```python
def BoxcarAVG(Points,n):
  # Let n be an odd number, 3 or 5 for example
  side = n // 2
  for i in range(side,len(Points) - side):
    Points[i][1] = sum([p[1] for p in Points[i-side:i+side+1]])/n
  return Points
```
